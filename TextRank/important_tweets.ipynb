```python
import pandas as pd
```

```python
df = pd.read_csv('Paragon.csv', index_col=0)
df
```

```python
def calculate_importance(tweet):
    quote_weight = tweet["quote"]
    reply_weight = tweet["reply"]
    retweet_weight = tweet["retweet"]
    like_weight = tweet["favorite"]
    text_rank_weight = tweet["textrank_score"]
    len_weight = tweet["len"]

    # formula based on priorities
    importance_score = 0.5 * text_rank_weight + 0.3 * retweet_weight + 0.1 * like_weight + 0.1 * len_weight + 0.1 * quote_weight + 0.1 * reply_weight

    return importance_score
```

```python
df["importance"] = df.apply(calculate_importance, axis=1)
```

```python
# Sort DataFrame by importance in descending order
sorted_df = df.sort_values(by="importance", ascending=False)
sorted_df
```

```python
# Select the top 10 tweets
top_10_tweets = sorted_df.head(10)

# Print the top 10 tweets
for i, row in top_10_tweets.iterrows():
    print(row['full_text'])
```
